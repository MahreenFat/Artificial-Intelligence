```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create a sample dataset
data = {
    "Age": [25, 30, 35, 40, 45, 50, 55, 60, 65, 70],
    "Salary": [40000, 50000, 60000, 70000, 80000, 90000, 100000, 110000, 120000, 130000],
    "Experience": [1, 3, 5, 7, 10, 12, 15, 18, 20, 25]
}
df = pd.DataFrame(data)

# Calculate descriptive statistics
mean_values = df.mean()
median_values = df.median()
std_deviation = df.std()

# Print descriptive statistics
mean_values, median_values, std_deviation
```

```python
# Visualize data distributions with histograms
plt.figure(figsize=(12, 8))

# Age histogram
plt.subplot(3, 1, 1)
plt.hist(df["Age"], bins=5, color='skyblue', edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")

# Salary histogram
plt.subplot(3, 1, 2)
plt.hist(df["Salary"], bins=5, color='lightgreen', edgecolor='black')
plt.title("Salary Distribution")
plt.xlabel("Salary")
plt.ylabel("Frequency")

# Experience histogram
plt.subplot(3, 1, 3)
plt.hist(df["Experience"], bins=5, color='salmon', edgecolor='black')
plt.title("Experience Distribution")
plt.xlabel("Experience (years)")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```
